```python
import serial # 导入串口库
import time #导入时间库

data_ser = serial.Serial("/dev/user_voice", 115200, timeout=5)  # 启动语音处理系统串口user_voice
data_ser1 = serial.Serial("/dev/user_arduino", 115200, timeout=5)  # 启动智能传感系统串口
data_ser.flushInput()  # 清空缓冲区

while True:
    data_count = data_ser.inWaiting() #获取串口缓冲区数据个数
    if data_count != 0:
        recv = data_ser.read(data_ser.in_waiting).decode("gbk") #读取串口数据，数据采用gbk编码
        result = recv.split()  #拆分字符串
        print(result[0])
        # 将语音处理系统发过来的指令转发给智能传感系统
        if (result[0] == "AA_kai-deng_BB"):  # 接收开灯指令，发送“A”
            data_ser1.write('A'.encode('utf-8'))
        elif (result[0] == "AA_guan-deng_BB"):  # 接收关灯指令，发送“B”
            data_ser1.write('B'.encode('utf-8'))
        elif (result[0] == "AA_kai-feng-san_BB"):  # 接收开风扇指令，发送“C”
            data_ser1.write('C'.encode('utf-8'))
        elif (result[0] == "AA_guan-feng-san_BB"):  # 接收开风扇指令，发送“D”
            data_ser1.write('D'.encode('utf-8'))
        elif (result[0] == "AA_xian-shi-wen-du_BB"): # 接收显示温湿度指令，发送“E”
            data_ser1.write('E'.encode('utf-8'))
```
