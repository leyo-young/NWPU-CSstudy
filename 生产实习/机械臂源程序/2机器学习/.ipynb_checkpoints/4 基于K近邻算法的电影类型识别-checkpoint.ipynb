```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
path='/home/retoo/Desktop/实验/数据集/2.机器学习/4-1 movies.xlsx'
movies_data = pd.read_excel(path)
print(movies_data)
movies_x = movies_data[["Action lens","Love lens"]]
movies_y = movies_data["target"]
x_train, x_test, y_train, y_test = train_test_split(movies_x, movies_y, test_size=0.2) #使用20%的数据来测试数据
knn = KNeighborsClassifier(n_neighbors=3)
movies_train = knn.fit(x_train,y_train)
movies_pred = movies_train.predict(x_test)
print(movies_pred)
print(movies_train.score(x_test,y_test))
```
