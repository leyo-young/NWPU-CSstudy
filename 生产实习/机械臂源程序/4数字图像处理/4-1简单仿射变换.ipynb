```python
# 导入库函数
import cv2
import numpy as np

# 读原图像
img=cv2.imread("/home/retoo/Desktop/实验/数据集/4.数字图像处理/4lena.jpg")

# 获取原图大小参数
height,width=img.shape[:2]

# 产生平移矩阵M,平移图像
x=100
y=200
M = np.float32([[1,0,x],[0,1,y]])
move=cv2.warpAffine(img,M,(width,height))

# 产生旋转矩阵M,旋转图像
N=cv2.getRotationMatrix2D((width/2,height/2),45,0.6)
rotate=cv2.warpAffine(img,N,(width,height))

# 显示原图像、平移图像、旋转图像
cv2.imshow("original",img)
cv2.imshow("move",move)
cv2.imshow("rotate",rotate)
cv2.waitKey()  # 等待按键
cv2.destroyAllWindows()  # 按键后关闭显示的所有窗口
```
