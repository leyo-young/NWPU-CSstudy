```python
# 
import cv2
%matplotlib inline
import matplotlib.pyplot as plt

#  
p1 = cv2.imread("/home/retoo/Desktop///4./1-1.jpg")
p2 = cv2.imread("/home/retoo/Desktop///4./1-2.jpg")
res1 = cv2.add(p1, p2)
res1_1 = p1 + p2
# 
res2 = cv2.addWeighted(p1, 0.3, p2, 0.7, 0)
# 
plt.figure(figsize=(128, 32))
plt.subplot(2, 3,  1)
plt.imshow(cv2.cvtColor(p1, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.subplot(2, 3,  2)
plt.imshow(cv2.cvtColor(p2, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.subplot(2, 3,  3)
plt.imshow(cv2.cvtColor(res1, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.subplot(2, 3,  4)
plt.imshow(cv2.cvtColor(res1_1, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.subplot(2, 3,  5)
plt.imshow(cv2.cvtColor(res2, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
