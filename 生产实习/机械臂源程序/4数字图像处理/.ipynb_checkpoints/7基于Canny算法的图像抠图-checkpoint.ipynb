```python
# 1.添加引用库
import cv2

# 2.对图像预处理
src = cv2.imread("/home/retoo/Desktop/实验/数据集/4 数字图像处理/7caomei.jpg")
blurred = cv2.GaussianBlur(src,(3,3),0)
gray = cv2.cvtColor(blurred,cv2.COLOR_BGR2GRAY)
ret,mask = cv2.threshold(gray,0,255,cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)

# 3.使用canny算子检测轮廓
grad_X = cv2.Sobel(mask,cv2.CV_16SC1,1,0)
grad_Y = cv2.Sobel(mask,cv2.CV_16SC1,0,1)
 # 将梯度值转化到8位上来
x_grad = cv2.convertScaleAbs(grad_X)
y_grad = cv2.convertScaleAbs(grad_Y)
edge = cv2.Canny(grad_X, grad_Y, 20, 100)

# 4.使用逻辑运算保留轮廓内容
# 用边缘做掩模，进行bitwise_and位运算
edge2 = cv2.bitwise_and(src,src, mask=mask)

#5.显示图像
cv2.imshow("original image",src)
cv2.imshow("Canny_edge", edge)
cv2.imshow("mask",mask)
cv2.imshow("bitwise_and", edge2)
cv2.waitKey()  # 等待按键
cv2.destroyAllWindows()  # 按键后关闭显示的所有窗口
```
