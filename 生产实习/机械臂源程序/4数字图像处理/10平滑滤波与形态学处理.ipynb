```python
# 导入库文件
import numpy as np
import cv2
```

```python
def blur(img):#均值滤波
    blur = cv2.blur(img, (5, 5))
    return blur
def GaussianBlur(img):#高斯滤波
    Guass = cv2.GaussianBlur(img, (5, 5), 0, 0)
    return Guass
def median(img):#中值滤波
    median = cv2.medianBlur(img,3)
    return median
def reosion(img):#腐蚀操作
    reo = cv2.erode(img, kernel, iterations=5)
    return reo
def dilate(img):#膨胀操作
    dilate = cv2.dilate(img,kernel,iterations=3)
    return dilate
def opening(img):#图像开运算
    opening = cv2.morphologyEx(img, cv2.MORPH_OPEN,kernel)
    return opening
def closing(img):#图像闭运算
    closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)
    return closing
```

```python
import cv2 as cv

img = cv.imread("/home/retoo/Desktop/实验/数据集/4.数字图像处理/10_2.png")  #读取图像
gauss = cv.GaussianBlur(img, (5, 5), 0, 0)  #通过5*5的卷积核进行高斯滤波
gray = cv.cvtColor(gauss, cv.COLOR_BGR2GRAY)  #转换为灰度图

dst = cv.threshold(gray, 70, 255, cv.THRESH_BINARY)[1]  #二值化图像
current = cv.erode(dst, None, iterations=1)  #对二值化的图像进行腐蚀操作
original_image = cv.resize(img, (640, 480))  #重新设置大小为（640，480）
current_image = cv.resize(current, (640, 480))
cv.imshow("original_image", original_image)  #显示图像
cv.imshow("current_image", current_image)
cv2.waitKey() #等待按键
cv2.destroyAllWindows()#按键后关闭刚显示的所有窗口
```
