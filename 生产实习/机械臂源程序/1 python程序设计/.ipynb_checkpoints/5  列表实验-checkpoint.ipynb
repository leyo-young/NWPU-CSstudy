```python
x=input("Please Input an integer less than 1000:")
x=eval(x)
t=x
i=2
#创建空列表，用于保存因子
result=[]
#这个while循环用来进行因式分解
while t>1:
    if t%i==0:
        result.append(i)
        t=t//i
    else:
        i=i+1
#显示因式分解的结果
print(x,'=',end='')
for i in range(len(result)):
    if i==len(result)-1:
        print(result[i],end='')
    else:
        print(result[i],'*',end='')

```
