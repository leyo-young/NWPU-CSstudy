```python
import re   #导入标准库re

tel_number = '''Suppose my Phone No. is 0606-1234666, 
                Yours number is 010-12345678,
                his number is 025-87993421.'''

pattern = re.compile(r'(\d{3,4})-(\d{7,8})')  # 逗号后面不能有空格
index=0
while True:
    matchresult = pattern.search(tel_number,index)
    if not matchresult:
        break
    print('-' * 30)
    print('success：')
    print('Search content:', matchresult.group(),'Start from:',matchresult.start(),'End at:',matchresult.end(),\
        'Its span is:',matchresult.span())
    index=matchresult.end()
```
