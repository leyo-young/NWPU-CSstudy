```python
import pickle
with open("student.dat","wb") as f:
    while True:
        id = int(input("请输入学号："))
        if id < 0:
           break
        name = input("请输入姓名：")
        sex = input("请输入性别：")
        age = int(input("请输入年龄："))
        print("**********")
        record=(id,name,sex,age)
        try:
            pickle.dump(record,f) #将一条学生信息写入f文件
        except:
            print("写入文件异常！")
with open("student.dat","rb") as f:
    result = []
    try:
        while True:
            record = pickle.load(f) #读取学生信息
            result.append(record)   #保存至result
    except EOFError:
            f.close()
result = sorted(result, reverse=False)
print(" 学号 ", "姓名  ", "性别 ", "年龄")
for record in result:
    print("%3d %4s %4s %4d"%(record[0],record[1],record[2],record[3]))



```
