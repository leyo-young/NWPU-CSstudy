```python
#（1）将 float 浮点型转化成 int 长整型
num_float = 3.5
print(num_float)
print(type(num_float))
num_int = int(num_float)
print(num_int)
print(type(num_int))
```

```python
#（2）将 2,3 转化为复数
num_complex = complex(2, 3)
print(num_complex)
print(type(num_complex))
```

```python
#（3）将字符串转化为数字类型
str_num = '789'
num = eval(str_num)
print(num)
print(type(num))
```

```python
#（4）整数与字符相互转换
print(chr(65))
print(chr(90))
print(chr(97))
print(chr(122))
print(ord('A'))
```

```python
#（5）显示布尔型变量及类型
print(True)  # True 首字母要大写
print(False)  # False 首字母要大写
print(type(True))  # 查看 True 的类型
print(type(False))  # 查看 False 的类型
```

```python
#(6)输入浮点型变量x,计算其一元二次方程的值
x=float(input('请输入浮点数：'))
y=x**2+2*x-10
print(y)
```

```python
#（7）将一个四位正整数反序输出
x=int(input('请输入一个四位正整数: '))
t1=x%10
t2=x%100//10
t3=x%1000//100
t4=x//1000
y=t4+t3*10+t2*100+t1*1000
print('{}的反序数是{}'.format(str(x),str(y)))
```
