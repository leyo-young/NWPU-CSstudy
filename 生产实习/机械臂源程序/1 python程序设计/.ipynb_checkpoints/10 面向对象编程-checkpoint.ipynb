```python
class Solution:     #定义类
    #定义计算网络延迟时间函数
    def networkDelayTime(self,times,N,K):
        # G[N+1][N+1]矩阵，用于保存节点间的花费时间，初始值为INF
        INF=0x3f3f3f3f
        G=[[INF for i in range(N+1)] for j in range(N+1)]
        #节点i至节点i本身的花费时间G[i][i]为0
        for i in range(1,N+1):
            G[i][i]=0
        #从times[]中提取不同节点间的花费时间，保存于 G[N+1][N+1]矩阵
        for i in range(0,len(times)):
            G[times[i][0]][times[i][1]]=times[i][2]
        #从矩阵G[N+1][N+1]提取k节点至其他节点的花费时间，保存于dis序列中
        dis=G[K][:]
        #vis[N+1]用于标记节点是否确定收到信息的最短花费时间，0：未确定，1：已确定，初始化为0
        vis=[0]*(N+1)
        #确定所有节点收到信息的最短花费时间，
        for i in range(0,N-1):
            Mini=INF
            p=K
            #从所有未确定最短花费时间的节点中，找到距离K节点最短花费时间的节点p
            for j in range(1,N+1):
                if vis[j]==0 and dis[j]<Mini:
                    Mini=dis[j]
                    p=j
            #将节点p标记为已经确定最短路花费时间。
            vis[p]=1
            #通过节点p更新所有节点距离K节点的最短花费时间；
            for j in range(1,N+1):
                if vis[j]==0 and dis[j]>dis[p]+G[p][j]:
                    dis[j]=dis[p]+G[p][j]
        #确定最长花费时间，即所有节点收到信息的最短花费时间的最大值，保存于ans中，返回
        ans=0
        for i in range(1,N+1):
            ans=max(ans,dis[i])
        if ans==INF:
            return -1
        return ans
if __name__=='__main__':
    solution=Solution()
    times=[[2,1,1],[2,3,1],[3,4,1],[2,5,3]]
    N=5
    K=2
    print("时间矩阵：",times)
    print("网络大小：", N)
    print("起始节点：", K)
    print("最长花费：", solution.networkDelayTime(times,N,K))
```
