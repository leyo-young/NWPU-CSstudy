```python
#输入集合，存放在列表中，方便操作，最后再转换成集合输出
list1=eval(input("请输入原集合，用列表表示："))
#用于存放集合所有的子集
sub_list_all=[]
# 产生2**N个子集，循环遍历0~2**N的每一个数，
N=len(list1)
for i in range(2**N):
# 用于存放每个单独循环中取出的子集
    sub_list=[]
    for j in range(len(list1)):
        if (i>>j)%2:
            sub_list.append(list1[j])
    sub_list_all.append(sub_list)
#输出原集合
print("集合为：",set(list1))
print("子集有",2**len(list1),"个，分别为：")
#将列表按元素的长度排序
sub_list_all.sort(key=len)
#以集合形式输出所有子集
for item in sub_list_all:
        if len(item)==0:
            print ('{}')
        else:
            print(set(item))
```
